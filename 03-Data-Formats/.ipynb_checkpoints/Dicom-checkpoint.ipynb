```python
path_to_head_mri = Path("SE000001/")
```

```python
list(path_to_head_mri.glob("*"))
```

```python
mri_data = []

for path in list(path_to_head_mri.glob("*")):
    data = pydicom.read_file(path)
    mri_data.append(data)
```

```python
for slice in mri_data[:5]:
    print(slice.SliceLocation)
```

```python
mri_data_sorted = sorted(mri_data, key=lambda slice: slice.SliceLocation)
```

```python
for slice in mri_data_sorted[:5]:
    print(slice.SliceLocation)
```

```python
full_volume = []
for slice in mri_data_sorted:
    full_volume.append(slice.pixel_array)
```

```python
fig, axis = plt.subplots(3, 3, figsize=(9,9))

slice_counter = 0

for i in range(3):
    for j in range(3):
        axis[i][j].imshow(full_volume[slice_counter], cmap="gray")
        slice_counter+=1
```

```python
import SimpleITK as sitk
```

```python
series_id = sitk.ImageSeriesReader.GetGDCMSeriesIDs(str(path_to_head_mri))
```

```python
series_file_names = sitk.ImageSeriesReader.GetGDCMSeriesFileNames(str(path_to_head_mri), series_id[0])
```

```python
series_reader = sitk.ImageSeriesReader()
series_reader.SetFileNames(series_file_names)
```

```python
image_data = series_reader.Execute()
```

```python
image_data.GetSize()
```

```python
image_data_numpy = sitk.GetArrayFromImage(image_data)
print(type(image_data_numpy))
print(image_data_numpy.shape)
```

```python
fig, axis = plt.subplots(3, 3, figsize=(9,9))

slice_counter = 0

for i in range(3):
    for j in range(3):
        axis[i][j].imshow(image_data_numpy[slice_counter], cmap="gray")
        slice_counter+=1
```
